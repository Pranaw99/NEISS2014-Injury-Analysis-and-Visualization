```python
# importing all the necessary library
import pandas as pd
import numpy as np
```

```python
# Reading the dataset into a dataframe alias df
df = pd.read_csv('NEISS2014.csv')
```

```python
#displaying the result
df.head()
```

```python
# renaming the column diag to Code to facilitate the joining
df.rename(columns={'diag':'Code'},inplace=True)
```

```python
df.head()
```

```python
df_disp = pd.read_csv('DiagnosisCodes.csv')
```

```python
#merging the dataset with the Diagnosis dataset
df_merge = pd.merge(df,df_disp)
```

```python
df_merge
```

```python
#disposition code 4 and 2 are of interest so filtered rows which satisfies any of the condition
df_filter = df_merge[(df_merge['disposition'] == 4) | (df_merge['disposition'] == 2)]
```

```python
df_filter.shape
```

```python
#count the injures frequencies on the above dataset
df_filter['Diagnosis'].value_counts()
```

```python
# I have also included the diag_other column in my analysis
df_filter['diag_other'].value_counts()
```

```python
# records with disposition column 6 are filtered out for my analysis
df_filter_2 = df_merge[(df_merge['disposition'] == 6)]
```
