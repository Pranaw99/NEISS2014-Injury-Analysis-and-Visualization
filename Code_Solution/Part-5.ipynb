```python
# importing all the necessary library
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# Reading the dataset into a dataframe alias df
df = pd.read_csv('NEISS2014.csv')
df_age= df[df['age']<110]
```

```python
plt.figure(figsize=(15,3))
sns.boxplot(x='diag',y='weight',data=df_age,color='red',palette='coolwarm')
```

```python
# Below few lines of code are my further exploring in a hope to get some useful insight
df_age_white = df_age[df_age['race'] == 'White']
sns.distplot(df_age_white['diag'],bins=10)
```

```python
sns.jointplot(x='age',y='prod1',data=df_age,kind= 'reg')
```

```python
sns.jointplot(x='weight',y='prod1',data=df_age,kind= 'reg')
```

```python
g = sns.FacetGrid(data=df_age,col='race',col_wrap=5,size=5)
g.map(plt.scatter,'diag','weight')
```

```python
# This code gets the distribution of weight based on individual diagnosis like 69,42,46,50
df_age_69 = df_age[df_age['diag'] == 69]
sns.distplot(df_age_69['weight'],bins=10)
```
